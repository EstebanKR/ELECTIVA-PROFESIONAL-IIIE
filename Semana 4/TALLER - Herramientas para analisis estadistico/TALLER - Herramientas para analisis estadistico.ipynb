```python
import numpy.testing as npt   #  necesario


def predecir(modelo, datos):
    X_train, X_test, y_train, y_test = datos
    if not hasattr(modelo, "positive"):
        modelo.positive = False

    if hasattr(modelo, "feature_names_in_") and isinstance(X_test, pd.DataFrame):
        req = list(modelo.feature_names_in_)
        faltantes = [c for c in req if c not in X_test.columns]
        if faltantes:
            raise ValueError(f"Faltan columnas requeridas por el modelo: {faltantes}")
        X_test = X_test[req]

    pred = modelo.predict(X_test)
    return np.asarray(pred)
```
